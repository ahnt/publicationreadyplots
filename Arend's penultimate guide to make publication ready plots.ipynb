```python
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * numpy.array(data)
    n = len(a)
    m, se = numpy.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
    return h
```

```python
C=mean_confidence_interval(Y, confidence=0.95)
figure(figsize=[7.5,3])
plot(Y.mean(0),'k')
xlabel("time",fontsize=16)
ylabel(r"$\bar{Y}$",fontsize=16)

xticks([0,100,200,300,400,500,600,700,800,900,1000],['0','100','200','300','400','500','600','700','800','900','1000'],fontsize=14)
yticks(fontsize=14)

X=linspace(0,1000,1000)

plot(Y.mean(0)+C,'r')
plot(Y.mean(0)-C,'r')
```
